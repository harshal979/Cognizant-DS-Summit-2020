```python
import random as rn
import pandas as pd
import numpy as np
```

```python
totalpopulation = 500
totalfitness=28
crossOver = 0.5
#mutationRate = 0.01
```

```python
populationlist= []
for outer in range(totalpopulation):
    set = []
    for inner in range(0,8):
        value = rn.randint(0,7)
        set.append(value)
    populationlist.append(set)
populationlist[1]
```

```python
def GetfitnessScore(data):
    set = data.copy()
    fitness = []
    while len(set)>0:
        fscore =len(set)-1
        queen = set.pop()
        qinrow = set.count(queen)
        fscore = fscore-qinrow
        pos = 1
        for i in reversed(set):
            if queen-pos==i:
                fscore-=1
            if queen+pos==i:
                fscore-=1
            pos+=1
        fitness.append(fscore)
    return(sum(fitness))
```

```python
fitnesslist = []
probability=[]
set = []
for outer in range(totalpopulation):
    set = populationlist[outer].copy()
    fitnessScore = GetfitnessScore(set)
    fitnesslist.append(fitnessScore)
    probability.append(fitnessScore/totalfitness)


#for fitness in fitnesslist:
    #probability.append(fitness/sum(fitnesslist))
    
Fitnessdataframe = pd.DataFrame({'chessboard':populationlist,'FitnessScore':fitnesslist,'Probability':probability})
Fitnessdataframe = Fitnessdataframe.sort_values(['FitnessScore'],ascending =False)
Fitnessdataframe = Fitnessdataframe.reset_index(drop=True)
```

```python
crossoverpoint = int(crossOver*8)
generationcount = 1000

for count in range(generationcount):
    totalpopulation = len(populationlist)
    draw = []
    draw.append(Fitnessdataframe[0:1]["chessboard"].values[0])
    draw.append(Fitnessdataframe[1:2]["chessboard"].values[0])
    draw.append(Fitnessdataframe[2:3]["chessboard"].values[0])

    #print(draw)
    if(GetfitnessScore(draw[0])==totalfitness | GetfitnessScore(draw[1])==totalfitness):
        #print('Generation ',count,' ',' Average Fitness Score ',round(Fitnessdataframe["FitnessScore"].mean(),4),' ', child1,' ',GetfitnessScore(child1),' ',child2,' ',GetfitnessScore(child2))
        print(draw[0],' ',draw[1])
        break
    child1 = draw[0][0:crossoverpoint]+draw[1][crossoverpoint:]
    child2 = draw[1][0:crossoverpoint]+draw[0][crossoverpoint:]
    child3 = draw[1][crossoverpoint:]+draw[2][0:crossoverpoint]
    child4 = draw[2][crossoverpoint:]+draw[1][0:crossoverpoint]
    
    child1[rn.randint(0,7)] = rn.randint(0,7)
    child2[rn.randint(0,7)] = rn.randint(0,7)
    #child3[rn.randint(0,7)] = rn.randint(0,7)
    child4[rn.randint(0,7)] = rn.randint(0,7)
    
    populationlist.append(child1)
    populationlist.append(child2)
    populationlist.append(child3)
    populationlist.append(child4)

    for outer in range(totalpopulation,len(populationlist)):
        fitnessScore = GetfitnessScore(populationlist[outer])
        fitnesslist.append(fitnessScore)
        probability.append(fitnessScore/totalfitness)

    #probability=[]
    #for fitness in fitnesslist:
        #probability.append(fitness/sum(fitnesslist))
        
    Fitnessdataframe = pd.DataFrame({'chessboard':populationlist,'FitnessScore':fitnesslist,'Probability':probability})
    Fitnessdataframe = Fitnessdataframe.sort_values(['FitnessScore'],ascending =False)
    Fitnessdataframe = Fitnessdataframe.reset_index(drop=True)
    print('Generation ',count,' ',' Average Fitness Score ',round(Fitnessdataframe["FitnessScore"].mean(),2),' ', child1,' ',GetfitnessScore(child1),' ',child2,' ',GetfitnessScore(child2))
```
